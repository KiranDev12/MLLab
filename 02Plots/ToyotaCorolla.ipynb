```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("ToyotaCorolla.csv")
#first 5 enteries
data.head() 
#last 5 enteries
data.tail()
#info summary
data.shape
```

```python
#trim data to 100 rows
limited_data = data.iloc[0:1000]
```

```python
#define co-ordinates
x = limited_data['KM']
y = limited_data['Price']
z = limited_data['Weight']

# SCATTER PLOT
plt.scatter(x, y, c=z, cmap="summer", edgecolors="grey", linewidth=1, alpha=0.75)

#colorbar below scatter plot
plt.colorbar(label="Weight", orientation="horizontal")

plt.title("Price of Toyota Corolla")
plt.xlabel('KM')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
# BOX PLOT
plt.boxplot([data['Price'], data['HP'], data['KM']])
plt.xticks([1,2,3], ["Price", "HP", "KM"])
plt.show()
```

```python
# HEAT MAP

sns.heatmap(data[['Price', 'KM', 'Weight']], cmap='jet')
```

```python
# COUNTOUR PLOT
x = data['KM']
y = data['Weight']
z = data['Price']

plt.tricontour(x,y,z, levels=20, cmap='jet')
plt.colorbar(label='Price')
plt.xlabel("KM")
plt.ylabel("Weight")
plt.title("Contour Plot")
```

```python
# 3-D PLOT
x = data['KM']
y = data['Doors']
z = data['Price']

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_trisurf(x, y, z, cmap='jet')

ax.set_xlabel('KM')
ax.set_ylabel('Doors')
ax.set_zlabel('Price')

plt.show()
```

```python
# HISTOGRAM
sns.distplot(data['cc'])
```

```python
# BAR PLOT
sns.countplot(x="Age_08_04", data=data)
```

```python
# Grouped BAR PLOT
sns.countplot(x="Price", data=data, hue='Automatic')
```
